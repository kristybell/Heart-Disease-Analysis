## Heart Disease Data

This code provides visualization of some data on heart disease.

```python
# Import Relevant Libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import heart disease Data from .csv file
heart_disease = pd.read_csv('heart_disease.csv')
heart_disease.head()                                  # Print the first 5 rows of the data
```

```python
# Create a histogram of age
heart_disease['age'].plot.hist(bins = 10);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]        # select the entries having an age > 50
over_50.head()                                            # print the top 5 rows of those with an age > 50
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10),
                               sharex = True);
# Change Plot Style
plt.style.use('seaborn-darkgrid')

# Add data to axis 'ax0'
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap = 'winter')

# Customize Plot using 'ax0' data
ax0.set(title = 'Heart Disease and Cholesterol Levels',
        ylabel = 'Cholesterol');

# Change the x axis limits (recall: ax0 and ax1 share the same axis)
ax0.set_xlim([50,80])

# Add a Legend to axis 'ax0' data plot
ax0.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontal line of the mean
ax0.axhline(y = over_50['chol'].mean(),
           linestyle = '--');


# Add data to axis 'ax1'
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap = 'winter')
# Customize Plot using 'ax1' data
ax1.set(title = 'Heart Disease and Max Heart Rate',
        xlabel = 'Age',
        ylabel = 'Max Heart Rate');

# Change y axis limit for ax1 data
ax1.set_ylim([60,200])

# Add a legend to axis 'ax1' data plot
ax1.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontal line of the mean
ax1.axhline(y = over_50['thalach'].mean(),
            linestyle = '--');

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize = 16,
             fontweight = 'bold');
```

```python
# Save the figure
fig.savefig('heart-disease-analysis-plot.png')
```
